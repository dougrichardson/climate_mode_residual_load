#### Renewable Energy Zones

```python
def get_REZ_mean(ds, gen_type, mask, region_code_name="SC2050"):
    """
    Compute values averaged for grid cells in each REZ
    
    ds: dataset of data to average
    gen_type: str, 'wind' or 'solar', used to set numeric codes for REZs to use
    mask: dataset of region mask
    region_code_name: str, append to 'region_code_' in ds e.g. 'SC2050'
    """
    if gen_type == "wind": 
        codes = [1, 3, 4]
    elif gen_type == "solar":
        codes = [2, 3]
    else:
        raise ValueError("Incorrect gen_type")
        
    da_list = []
    for region in mask["region"].values:
        m = mask.sel(region=region) # Mask for this region
        
        if m["region_code_"+region_code_name] in codes: # Skip regions without this gen_type
            d = ds.where(m, drop=True).mean(["lat", "lon"])
            d = d.expand_dims({"region": [region]})
            da_list.append(d)
        else:
            pass
        
    return xr.concat(da_list, dim="region")
```

## Relationships with capacity factors

```python
def correlation_by_month(ds1, ds2):
    """
    Return dataset with correlations computed for each month
    
    ds1, ds2: datasets to correlate
    """
    cor_list = []
    pval_list = []
    for m in range(1, 13):
        ds1_m = fn.sel_month(ds1, m)
        ds2_m = fn.sel_month(ds2, m)
        
        # ds1_m = ds1_m.dropna("time")
        # ds2_m = ds2_m.dropna("time")
        
        # cor = fn.xr_spearmanr(ds1_m, ds2_m).expand_dims({"month": [m]})
        _cor, _pval = fn.xr_spearmanr(ds1_m, ds2_m)

        _cor = _cor.expand_dims({"month": [m]})
        _pval = _pval.expand_dims({"month": [m]})

        cor_list.append(_cor)
        pval_list.append(_pval)
        
    return xr.concat(cor_list, dim="month"), xr.concat(pval_list, dim="month")
```

```python
def get_correlation_by_month(cf_ds_dict, mode_dict, compute=False):
    """
    Compute and write, or load, correlation between capacity factors and modes.
    
    cf_ds_dict, mode_dict: dict of capacity factors and mode data.
    compute: bool, whether to compute or load.
    """
    import pickle
    
    fp = "/g/data/w42/dr6273/work/projects/Aus_energy/monthly_data/"
    
    if compute:
        cf_cor = {}
        cf_pval = {}
        
        for cf_ds, cf_name in zip(
            list(cf_ds_dict.values()),
            list(cf_ds_dict.keys())
        ):
            cf_cor[cf_name] = {}
            cf_pval[cf_name] = {}

            for mode_ds, mode_name in zip(
                list(mode_dict.values()),
                list(mode_dict.keys())
            ):
                cf_cor[cf_name][mode_name] = {}
                cf_pval[cf_name][mode_name] = {}
                
                cf_cor[cf_name][mode_name]["monthly"], cf_pval[cf_name][mode_name]["monthly"] = correlation_by_month(cf_ds, mode_ds)
                cf_cor[cf_name][mode_name]["annual"], cf_pval[cf_name][mode_name]["annual"] = fn.xr_spearmanr(cf_ds, mode_ds)
                
        with open(fp + "capacity_factor_climate_mode_correlation.pkl", "wb") as handle:
            pickle.dump(cf_cor, handle, protocol=pickle.HIGHEST_PROTOCOL)

        with open(fp + "capacity_factor_climate_mode_correlation_pval.pkl", "wb") as handle:
            pickle.dump(cf_pval, handle, protocol=pickle.HIGHEST_PROTOCOL)

        return cf_cor, cf_pval
    else:
        with open(fp + "capacity_factor_climate_mode_correlation.pkl", "rb") as handle:
            cor = pickle.load(handle)
        with open(fp + "capacity_factor_climate_mode_correlation_pval.pkl", "rb") as handle:
            pval = pickle.load(handle)
        return cor, pval
```

```python
def adjust_lightness(color, amount=0.5):
    """
        Adjusts lightness of the given color by multiplying (1-luminosity) by the given amount.
        Input can be matplotlib color string, hex string, or RGB tuple.

        Examples:
        >> adjust_lightness('g', 0.3)
        >> adjust_lightness('#F034A3', 0.6)
        >> adjust_lightness((.3,.55,.1), 0.5)
        
        From: https://stackoverflow.com/a/49601444
    """
    import matplotlib.colors as mc
    import colorsys
    try:
        c = mc.cnames[color]
    except:
        c = color
    c = colorsys.rgb_to_hls(*mc.to_rgb(c))
    return colorsys.hls_to_rgb(c[0], max(0, min(1, amount * c[1])), c[2])
```

```python
def rez_scatter(ax, lat_lon, cor_da, gen_type, mask, fill_rez=None, region_code_name="SC2050"):
    """
    Scatter REZ correlation
    
    ax: axes
    lat_lon: str, 'lat' or 'lon', or None, indicating which dimension to average over.
    cor_da, dataset, array of correlations
    gen_type: str, 'wind' or 'solar'
    mask: dataset of REZs
    fill_rez: list of str, indicating REZ IDs for which to plot filled markers
    region_code_name: str, append to 'region_code_' in ds e.g. 'SC2050'
    """
    on_c = "tab:blue"
    off_c = "tab:blue"
    ms = 20
    
    d = get_REZ_mean(cor_da, gen_type, mask)
    d_on = d.where(d["region_code_"+region_code_name] != 4, drop=True)
    d_off = d.where(d["region_code_"+region_code_name] == 4, drop=True)
    if lat_lon == "lon":
        ax.scatter(d_on.values, d_on.latitude.values, marker="s", color=on_c, s=ms, ec="white")
        ax.scatter(d_off.values, d_off.latitude.values, marker="x", color=off_c, s=ms, ec="white")
    elif lat_lon == "lat":
        ax.scatter(d_on.longitude.values, d_on.values, marker="s", color=on_c, s=ms, ec="white")
        ax.scatter(d_off.longitude.values, d_off.values, marker="x", color=off_c, s=ms, ec="white")
    else:
        ax.scatter(d_on.longitude.values, d_on.latitude.values, marker="o", color="None", ec="k", s=ms*1.5+0.9, lw=1, zorder=1)
        ax.scatter(d_off.longitude.values, d_off.latitude.values, marker="s", color="None", ec="k", s=ms*2.5+0.9, lw=1, zorder=1)
        
        ax.scatter(d_on.longitude.values, d_on.latitude.values, marker="o", color="None", ec=on_c, s=ms*1.5, lw=1, zorder=2)
        ax.scatter(d_off.longitude.values, d_off.latitude.values, marker="s", color="None", ec=off_c, s=ms*2.5, lw=1, zorder=2)
        
        if isinstance(fill_rez, list):
            for r in fill_rez:
                _d = d.sel(region=r)
                if _d["region_code_"+region_code_name] != 4:
                    ax.scatter(_d.longitude.values, _d.latitude.values, marker="o", color=on_c, s=ms*1.5, zorder=2)
                else:
                    ax.scatter(_d.longitude.values, _d.latitude.values, marker="s", color=off_c, s=ms*2.5, zorder=2)
```

```python
def _add_contour(ax, da, pvals):
    """ add pvalue contour """
    mask = pvals.astype(float)

    ax.contour(
        da[da.dims[1]],
        da[da.dims[0]],
        mask,
        levels=[0.5],
        colors="cyan",
        linewidths=1,
        zorder=10,
        transform=ccrs.PlateCarree()
    )
```

```python
def max_month(ds):
    """
    Return array of month and value of the maximum absolute value of ds
    """
    idx = abs(ds).idxmax("month")
    c = ds.sel(month=idx)
    return idx, c
```

```python
def panel_month_cor(gen_type, cor_vmax=0.7, cmap_season=False):
    """
    Plot max correlation and corresponding month for each mode
    
    gen_type: str, 'wind' or 'solar'
    cor_vmax: float, colormap maximum
    cmap_season: bool, whether to plot month or season of max correlation
    """
    ### Custom colorbar for month
    if cmap_season:
        color_sel_list = [0, 1, 2, 4]
        lightness_adj = [1.2] * 3
        cbar_ticks = np.arange(1.5, 11, 3)
        cbar_ticklabels = ["DJF", "MAM", "JJA", "SON"]
        time_name = "Season"
    else:
        color_sel_list = [7,4,0,1]
        lightness_adj = [0.9, 1.2, 1.5]
        cbar_ticks = np.arange(0.5, 12, 1)
        cbar_ticklabels = ["Dec", "", "Feb", "", "Apr", "", "Jun", "", "Aug", "", "Oct", ""]
        time_name = "Month"
        
    # select desired starting colours
    colours_sel = np.asarray(plt.rcParams['axes.prop_cycle'].by_key()['color'])[color_sel_list]
    # get three grads of lightness
    l_colours = [adjust_lightness(c, a) for c in colours_sel for a in lightness_adj]
    # set up cmap and norm
    cmap = matplotlib.colors.ListedColormap(l_colours)
    norm = matplotlib.colors.BoundaryNorm(np.arange(0, 12.01), cmap.N) 

    fig, ax = plt.subplots(2, 3, figsize=(6., 3), subplot_kw={"projection": ccrs.PlateCarree()})

    for a in ax.flatten():
        a.coastlines()

    for i, (mode_ds, mode_name) in enumerate(zip(
        [nino34, dmi, sam],
        ["ENSO", "IOD", "SAM"]
    )):
        idx, c = max_month(cf_cor[gen_type][mode_name]["monthly"])
        
        # Set month=12 to 0, so we can plot DJF together
        idx = idx.where(idx != 12, 0)
        
        p_c = c.plot(
            ax=ax[0,i],
            add_colorbar=False,
            vmax=cor_vmax,
            vmin=-1 * cor_vmax,
            cmap="PiYG"
        )
        p_m = idx.plot(
            ax=ax[1,i],
            levels=range(12),
            cmap=cmap,
            norm=norm,
            add_colorbar=False
        )
        print(abs(c).max().values)
        
        # # REZ boundaries
        # if gen_type == "wind":
        #     codes = [1, 3, 5, 7, 8]
        # else:
        #     codes = [2, 3, 6, 7]
        # for j in range(len(rez_gpd)):
        #     d = rez_gpd.iloc[j:j+1]
        #     code = d.iloc[0]['code']
        #     if code in codes:
        #         d.plot(ax=ax[0,i], alpha=1, edgecolor="tab:blue", facecolor="None", label='Wind', rasterized=True)
        #         d.plot(ax=ax[1,i], alpha=1, edgecolor="white", facecolor="None", label='Wind', rasterized=True)

    
    ax[0,0].set_title("a) ENSO-"+gen_type, fontsize=10, loc="left")
    ax[0,1].set_title("b) IOD-"+gen_type, fontsize=10, loc="left")
    ax[0,2].set_title("c) SAM-"+gen_type, fontsize=10, loc="left")
    
    ax[1,0].set_title("d) ENSO-"+gen_type, fontsize=10, loc="left")
    ax[1,1].set_title("e) IOD-"+gen_type, fontsize=10, loc="left")
    ax[1,2].set_title("f) SAM-"+gen_type, fontsize=10, loc="left")
        
    ax[0,0].text(-0.2, 0.5, r"$\max{\ |\rho|}$", va="center", rotation=90, transform=ax[0,0].transAxes)
    ax[1,0].text(-0.2, 0.5, time_name + r" of $\max{\ |\rho|}$", va="center", rotation=90, transform=ax[1,0].transAxes)
    
    plt.subplots_adjust(wspace=0.0, hspace=0.25)
    
    cax1 = fig.add_axes([0.9, 0.53, 0.01, 0.35])
    cbar1 = fig.colorbar(p_c, cax=cax1, ticks=np.arange(-1 * cor_vmax, cor_vmax + 0.01, cor_vmax), orientation="vertical")
    cbar1.ax.set_ylabel(r"Spearman $\rho$ [-]")
    
    cax2 = fig.add_axes([0.9, 0.11, 0.01, 0.35])
    cbar2 = fig.colorbar(p_m, cax=cax2, orientation="vertical")
    cbar2.set_ticks(cbar_ticks)
    cbar2.ax.minorticks_off()
    cbar2.set_ticklabels(cbar_ticklabels)
    cbar2.ax.set_ylabel(time_name)
```

### Capacity factors averaged by lat and lon

```python
def dim_cor(dim, cor_da, site_da):
    """
    Return correlations averaged by lat or lon, for all grid cells
    and for those with farms only.
    
    dim: str, 'lat', or 'lon'
    cor_da: dataset, correlations array to average
    site_da: dataset, array indicating which grid cells have sites
    """
    cor_all = cor_da.mean(dim)
    cor_site = cor_da.where(site_da > 0).mean(dim)
    return cor_all, cor_site
```

```python
def plot_lat_lon(ax, lat_lon, cor_da, site_da, color, mask=None, p1=0.0, p2=1):
    """
    Plot cor_da averaged over lat or lon
    
    ax: axes
    lat_lon: str, 'lat' or 'lon' indicating which dimension to average over
    cor_da, dataset, array of correlations
    site_da: dataset, array indicating which grid cells have sites
    color: str
    mask: dataset or None, array to use as a mask for cor_da
    """
    if mask is not None:
        cor_da = cor_da.where(mask==1)
        
    # percentile curves
    p1_c = cor_da.quantile(p1, lat_lon)
    p2_c = cor_da.quantile(p2, lat_lon)
    
    # Mean correlation for all cells and site cells
    cor_all, cor_site = dim_cor(lat_lon, cor_da, site_da)
    
    if lat_lon == "lat":
        ax.fill_between(p1_c["lon"], p1_c, p2_c, color=color, alpha=0.3)
        cor_all.plot(ax=ax, x="lon", color=color)
        # ax.scatter(cor_site["lon"], cor_site, color="gray", s=12)
        ax.axhline(0, color="k", lw=0.5)
        ax.set_ylim(-0.7, 0.7)
    else:
        ax.fill_betweenx(p1_c["lat"], p1_c, p2_c, color=color, alpha=0.3)
        cor_all.plot(ax=ax, y="lat", color=color)
        # ax.scatter(cor_site, cor_site["lat"], color="gray", s=12)
        ax.axvline(0, color="k", lw=0.5)
        ax.set_xlim(-0.7, 0.7)

    ax.set_title("")
```

```python
def rez_scatter(ax, lat_lon, cor_da, gen_type, mask):
    """
    Scatter REZ correlation
    
    ax: axes
    lat_lon: str, 'lat' or 'lon' indicating which dimension to average over
    cor_da, dataset, array of correlations
    gen_type: str, 'wind' or 'solar'
    mask: dataset of REZs
    """
    d = get_REZ_mean(cor_da, gen_type, mask)
    d_on = d.where(~d.region.isin(offshore_rez), drop=True)
    d_off = d.where(d.region.isin(offshore_rez), drop=True)
    if lat_lon == "lon":
        # ax.scatter(d.values, d.latitude.values, marker="^", color="r", s=12)
        ax.scatter(d_on.values, d_on.latitude.values, marker="^", color="r", s=12)
        ax.scatter(d_off.values, d_off.latitude.values, marker="x", color="tab:blue", s=12)
    else:
        # ax.scatter(d.longitude.values, d.values, marker="^", color="r", s=12)
        ax.scatter(d_on.longitude.values, d_on.values, marker="^", color="r", s=12)
        ax.scatter(d_off.longitude.values, d_off.values, marker="x", color="tab:blue", s=12)
```
